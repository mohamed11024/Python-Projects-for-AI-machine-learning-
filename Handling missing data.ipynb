```python
import pandas as pd
df = pd.read_csv("C:\\Users\\Mo\\Downloads\\chapter3_assets\\chapter_3_assets\\4_handling_missing_data_fillna_dropna_interpolate\\weather_data.csv",parse_dates = ["day"])
df
```

```python
df.set_index('day',inplace=True)
df
```

```python
df.fillna(0)
df
```

```python
df.fillna({
    'temperature' : df.temperature.mean(),
    'windspeed' : df.windspeed.mean()
})
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(method='bfill',axis='columns')
```

```python
df.fillna(method = 'ffill',limit = 1)
```

```python
df.interpolate()
```

```python
df.dropna()
```

```python
df.dropna(how="any")
```

```python
df.dropna(how="all")
```
