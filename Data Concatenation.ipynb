```python
import pandas as pd
india_weather = pd.DataFrame({
    'city':['Mumbai','Delhi','Banglore'],
    'temperature':[32,45,30],
    'humidity':[80,60,78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    'city':['Newyork','Chicago','Orlando'],
    'temperature':[21,14,35],
    'humidity':[68,65,75]
})
us_weather
```

```python
df = pd.concat([india_weather,us_weather],keys = ["indea","us"])
df
```

```python
df.loc["indea"]
```

```python
df.loc["us"]
```

```python
temperature_df = pd.DataFrame({
    'city':['Mumbai','Delhi','Banglore'],
    'temperature':[32,45,30]
},index = [0,1,2])

temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'city':['Delhi','Mumbai'],
    'windspeed':[7,12]
},index = [1,0])

windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df],axis = 1)
df
```
