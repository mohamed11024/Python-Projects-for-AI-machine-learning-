```python
import pandas as pd
#for movies.csv
df_movies = pd.read_csv("C:\\Users\\Mo\\Downloads\\chapter3_assets\\chapter_3_assets\\7_concat_merge\\Exercise\\movies.csv")
#for financials.csv
df_financials = pd.read_csv("C:\\Users\\Mo\\Downloads\\chapter3_assets\\chapter_3_assets\\7_concat_merge\\Exercise\\financials.csv")
#for languages.csv
df_languages = pd.read_csv("C:\\Users\\Mo\\Downloads\\chapter3_assets\\chapter_3_assets\\7_concat_merge\\Exercise\\languages.csv")
```

```python
#read the csv file: 'new_movies.csv'
df_new_movies = pd.read_csv("C:\\Users\\Mo\\Downloads\\chapter3_assets\\chapter_3_assets\\7_concat_merge\\Exercise\\new_movies.csv")
#show the dataframe
df_new_movies
```

```python
#concat the two dataframes ['df_movies', df_new_movies]
df_movies = pd.concat([df_movies,df_new_movies],ignore_index = True)
#show the last 5 rows
df_movies.head(5)
```

```python
#perform the join operation
df_movies = pd.merge(df_movies, df_languages, on = 'language_id', how = 'inner')

#show the top 5 rows
df_movies.head()
```

```python
#perform the join operation
df_movies = pd.merge(df_movies, df_financials, on = 'movie_id', how = 'left')

#show the last 5 rows
df_movies.tail()
```

```python
#save the dataframe to 'final_complete_data.csv'
df_movies.to_csv("C:\\Users\\Mo\\Downloads\\chapter3_assets\\chapter_3_assets\\7_concat_merge\\Exercise\\final_complete_data.csv", columns = ['movie_id', 'title', 'lang_name', 'budget', 'revenue', 'currency'], 
                 index = False)
```
